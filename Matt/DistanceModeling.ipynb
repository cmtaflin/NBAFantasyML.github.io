```python
# Dependencies and Setup
import json
import pandas as pd
import requests
import numpy as py
```

```python
#load the csv files
venues = "SportsLocations_limited_working2.csv"
dist_venue = "DistanceMiles_Final.csv"

# Read File and store into Pandas data frame
venue_df = pd.read_csv(venues)
distance_df = pd.read_csv(dist_venue)
```

```python
#popular_item_df = pop_item_df[["Item ID", "Item Name_x", "Purchase ID_y", "Price_x", "Price"]]
```

```python
venue_small = venue_df[['VENUEID','NAME','LATITUDE','LONGITUDE','Lat_Y','Long_X','TEAM','POPULATION']]
venue_small.head()
```

```python
venue_final = venue_small[['VENUEID','NAME','LATITUDE','LONGITUDE','TEAM','POPULATION']]
venue_final.head()
```

```python
distance_new = distance_df.rename(columns={"VenueID":"VENUEID"})
distance_new.head()
```

```python
team_locations = pd.merge(venue_final, distance_new, on="VENUEID")
team_locations.head()
```

```python
CityList_df = pd.read_csv("teamCityList.csv")
```

```python
CityList_df.head()
```

```python
team_locations_final = pd.merge(team_locations, CityList_df, on="VENUEID")
team_locations_final.head()
```

```python
# Load NBA_TEAMS table
#export to csv
team_locations_final.to_csv("NBA_Teams_Locations.csv", index=False)
```
