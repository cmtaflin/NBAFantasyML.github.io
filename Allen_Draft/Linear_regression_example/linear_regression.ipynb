```python
# dependencies
import pandas as pd
```

```python
# read excel file into dataframes
file_to_load = "jrue_holiday.xlsx"
train = pd.read_excel(file_to_load)
train.head()
```

```python
X = train[["MP", "TOV%", "USG%", "OpSTL"]]
y = train["TOV"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
# Use train_test_split to create training and testing data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create the model using LinearRegression

from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Fit the model to the training data and calculate the scores for the training and testing data

model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
# read excel file into dataframes
file_to_load = "next_week_games.xlsx"
next_week = pd.read_excel(file_to_load)
test = next_week[["MP", "TOV%", "USG%", "OpSTL"]]
test
```

```python
prediction = model.predict(test)
prediction
```

```python
next_week["TOV"] = prediction
```
